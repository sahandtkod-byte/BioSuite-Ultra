# BioSuite Ultra - Sequence Analysis

This notebook demonstrates the core sequence analysis features:
- FASTA/FASTQ I/O
- GC content, reverse complement, translation
- ORF finding, primer design, restriction enzyme mapping
- Codon usage analysis

```python
import sys
sys.path.insert(0, '..')

from biosuite.core.sequence import (
    read_fasta, gc_content, reverse_complement, translate, sequence_stats
)
from biosuite.core.orf_finder import find_orfs, find_restriction_sites, design_primers
from biosuite.core.codon_usage import codon_usage_table, kmer_composition, sequence_complexity
```

## 1. Basic Sequence Analysis

```python
# Example DNA sequence
seq = "ATGAAATTTTAAATGCGCGCGATCGATCGATCGATCGAATTTTAA"

# Sequence statistics
stats = sequence_stats(seq)
print("Sequence Statistics:")
for k, v in stats.items():
    print(f"  {k}: {v}")
```

```python
# GC content
gc = gc_content(seq)
print(f"GC Content: {gc:.2f}%")

# Reverse complement
rc = reverse_complement(seq)
print(f"Original:    {seq}")
print(f"Rev Comp:    {rc}")

# Translation
protein = translate(seq, frame=1)
print(f"Protein:     {protein}")
```

## 2. ORF Finding

```python
# Find open reading frames
long_seq = "TTTATGAAATTTTAATTTATGCCCAAAGGGTTTATGAAAAAAATTTTTTAA"
orfs = find_orfs(long_seq, min_length=1)

print(f"Found {len(orfs)} ORFs:")
for orf in orfs:
    print(f"  Frame {orf.frame}: pos {orf.start}-{orf.end}, "
          f"len={orf.length}, protein={orf.protein[:20]}...")
```

## 3. Restriction Enzyme Mapping

```python
# Find restriction sites
dna = "GAATTCATGAAATTTGAATTCATGAAAAAAGAATTC"
sites = find_restriction_sites(dna, enzymes=['EcoRI'])

print(f"Found {len(sites)} EcoRI sites:")
for site in sites:
    print(f"  Position {site.position}: {site.recognition_site}")
```

## 4. Primer Design

```python
# Design PCR primers
template = "ACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGTACGT" * 3
fwd, rev = design_primers(template)

if fwd and rev:
    print(f"Forward primer: {fwd.sequence}")
    print(f"  Tm: {fwd.tm:.1f}C, GC: {fwd.gc_content:.1f}%")
    print(f"Reverse primer: {rev.sequence}")
    print(f"  Tm: {rev.tm:.1f}C, GC: {rev.gc_content:.1f}%")
```

## 5. K-mer Composition & Sequence Complexity

```python
# K-mer composition
kmers = kmer_composition(seq, k=3)
print("Top 3-mers:")
for kmer, count in sorted(kmers.items(), key=lambda x: -x[1])[:10]:
    print(f"  {kmer}: {count}")

# Sequence complexity
complexity = sequence_complexity(seq)
print(f"\nAverage complexity: {complexity['average_complexity']:.3f}")
print(f"Low complexity: {complexity['is_low_complexity']}")
```
