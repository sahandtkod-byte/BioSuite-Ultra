# BioSuite Ultra - GWAS & Epitope Prediction

This notebook demonstrates:
- GWAS analysis with simulated data
- T-cell and B-cell epitope prediction

```python
import sys
sys.path.insert(0, '..')

from biosuite.core.gwas import run_gwas, detect_lead_snps, generate_gwas_data, format_gwas_report
from biosuite.core.epitope import (
    predict_t_cell_epitopes, predict_b_cell_epitopes,
    predict_linear_epitopes, format_epitope_report
)
```

## 1. GWAS Analysis

```python
# Generate simulated GWAS data
data = generate_gwas_data(n_snps=1000, seed=42)
print(f"Generated {len(data)} SNPs")
print(data.head())
```

```python
# Run GWAS
results = run_gwas(data)
leads = detect_lead_snps(results)

print(format_gwas_report(results, leads))
```

## 2. Epitope Prediction

```python
# Protein sequence for epitope prediction
protein = "MKWVTFISLLFLFSSAYSRGVFRRDAHKSEVAHRFKDLGEENFKALVLIAFAQYLQQCPFEDHVKLVNEVTEFAKTCVADESAENCDKS"

# T-cell epitopes (MHC binding)
t_cell = predict_t_cell_epitopes(protein, mhc_type="A0201")
print(f"T-cell epitopes found: {len(t_cell)}")
for e in t_cell[:5]:
    print(f"  {e.peptide} (score={e.score:.3f}, pos={e.start}-{e.end})")
```

```python
# B-cell epitopes (surface accessibility)
b_cell = predict_b_cell_epitopes(protein)
print(f"B-cell epitopes found: {len(b_cell)}")
for e in b_cell[:5]:
    print(f"  {e.peptide} (score={e.score:.3f}, pos={e.start}-{e.end})")
```

```python
# Full report
print(format_epitope_report(t_cell, b_cell, "HSA Protein"))
```
