# BioSuite Ultra - Alignment & Phylogenetics

This notebook demonstrates:
- Pairwise alignment (Needleman-Wunsch, Smith-Waterman)
- Multiple sequence alignment
- Phylogenetic tree construction

```python
import sys
sys.path.insert(0, '..')

from biosuite.core.alignment import needleman_wunsch, smith_waterman
from biosuite.core.msa import auto_align, consensus_sequence
from biosuite.core.phylogeny import distance_matrix, upgma_tree
from biosuite.core.blast import build_kmer_index, search_index
```

## 1. Pairwise Alignment

```python
# Global alignment (Needleman-Wunsch)
seq1 = "AGTACGCA"
seq2 = "TATGC"

aln1, aln2, score = needleman_wunsch(seq1, seq2)
print("Global Alignment (Needleman-Wunsch):")
print(f"  {aln1}")
print(f"  {aln2}")
print(f"  Score: {score}")
```

```python
# Local alignment (Smith-Waterman)
seq1 = "CCATCGTACG"
seq2 = "GGTACGA"

aln1, aln2, score = smith_waterman(seq1, seq2)
print("Local Alignment (Smith-Waterman):")
print(f"  {aln1}")
print(f"  {aln2}")
print(f"  Score: {score}")
```

## 2. Multiple Sequence Alignment

```python
# Progressive MSA
sequences = [
    "ATCGATCG",
    "ATCAATCG",
    "ATCGATAG",
    "ATCGATGG",
]

result = auto_align(sequences)
consensus = consensus_sequence(result)

print("Multiple Sequence Alignment:")
for name, seq in result.alignment.items():
    print(f"  {name}: {seq}")
print(f"\nConsensus: {consensus}")
```

## 3. Phylogenetic Tree

```python
# Build distance matrix and UPGMA tree
seqs = ["ATCGATCG", "ATCAATCG", "GGGGCCCC", "AAAATTTT"]
names = ["Human", "Chimp", "Mouse", "Yeast"]

dist = distance_matrix(seqs)
print("Distance Matrix:")
print(dist)

# Build UPGMA tree
tree = upgma_tree(dist, names)
print(f"\nTree: {tree}")
```

## 4. BLAST-like Search

```python
# Build k-mer index and search
database = {
    "seq1": "ATCGATCGATCGATCG",
    "seq2": "GCTAGCTAGCTAGCTA",
    "seq3": "TTTTAAAACCCCGGGG",
}

index = build_kmer_index(database, k=4)
hits = search_index("ATCGATCG", index, k=4)

print(f"Search hits: {len(hits)}")
for hit in hits:
    print(f"  {hit}")
```
