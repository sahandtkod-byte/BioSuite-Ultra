# BioSuite Ultra - Pipeline & Batch Processing

This notebook demonstrates:
- Building and running analysis pipelines
- Batch processing multiple samples
- Generating HTML reports

```python
import sys
sys.path.insert(0, '..')

from biosuite.core.workflow.pipeline import Pipeline
from biosuite.core.workflow.batch import BatchProcessor
from biosuite.core.workflow.report import HTMLReport
from biosuite.core.sequence import gc_content, reverse_complement, translate
```

## 1. Pipeline Builder

```python
# Create a sequence analysis pipeline
p = Pipeline("DNA Analysis Pipeline")

# Add steps
p.add_step("gc", gc_content, args=("ATGCGCGCGCGCGCGC",))
p.add_step("revcomp", reverse_complement, args=("ATGCGCGCGCGCGCGC",))
p.add_step("translate", translate, args=("ATGCGCGCGCGCGCGC",))

# Run the pipeline
p.run()

# View results
print(p.summary())
```

## 2. Batch Processor

```python
# Process multiple sequences in parallel
sequences = [
    "ATCGATCG",
    "GGGGCCCC",
    "AAAATTTT",
    "GCGCGCGC",
    "TTTTAAAA",
]

bp = BatchProcessor("GC Content Batch")
bp.add_samples(sequences, lambda seq: gc_content(seq))
bp.run(max_workers=2)

print(bp.summary())
print(f"\nResults: {bp.get_results()}")
```

## 3. HTML Report Generation

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

# Create a report
report = HTMLReport("BioSuite Analysis Report", "Generated from notebook")

# Add statistics
report.add_stats({"Sequences": 5, "Pipeline Steps": 3, "GC Average": 50.0})

# Add a plot
fig, ax = plt.subplots()
ax.bar(["Seq1", "Seq2", "Seq3", "Seq4", "Seq5"], [50, 100, 0, 100, 0])
ax.set_ylabel("GC Content %")
ax.set_title("GC Content Comparison")
report.add_plot(fig, "GC Content Bar Chart")

# Add text
report.add_text("All sequences analyzed successfully.", "Summary")

# Save
report.save("analysis_report.html")
print("Report saved to analysis_report.html")
```
