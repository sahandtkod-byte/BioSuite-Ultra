# BioSuite Ultra - Visualization Gallery

This notebook demonstrates the 36+ visualization types available in BioSuite.

```python
import sys
sys.path.insert(0, '..')
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np

from biosuite.plotting.upset_plots import plot_upset
from biosuite.plotting.conservation_plots import plot_logo_with_conservation
from biosuite.plotting.synteny import plot_synteny_dotplot
```

## 1. UpSet Plot (Multi-Set Intersections)

```python
# Create UpSet plot
sets = {
    'DE Genes': {'BRCA1', 'TP53', 'MYC', 'EGFR', 'PTEN'},
    'Cancer Genes': {'TP53', 'MYC', 'RB1', 'APC'},
    'Drug Targets': {'BRCA1', 'EGFR', 'PTEN', 'RB1'},
}

fig = plot_upset(sets, title="Gene Overlap Analysis")
plt.savefig('upset_plot.png', dpi=120, bbox_inches='tight')
plt.close()
print("Saved: upset_plot.png")
```

## 2. Sequence Logo

```python
# Create sequence logo with conservation
sequences = [
    "ACGTACGTACGT",
    "ACGAACGTACGT",
    "ACGTACGTACGA",
    "ACGTACGTACGT",
    "ACGTACGTACGT",
]

fig = plot_logo_with_conservation(sequences)
plt.savefig('sequence_logo.png', dpi=120, bbox_inches='tight')
plt.close()
print("Saved: sequence_logo.png")
```

## 3. Synteny Dotplot

```python
# Create synteny dotplot
genes1 = ['GeneA', 'GeneB', 'GeneC', 'GeneD', 'GeneE']
genes2 = ['GeneA', 'GeneC', 'GeneB', 'GeneE', 'GeneD']

fig = plot_synteny_dotplot(genes1, genes2,
                           genome1_name="Species 1",
                           genome2_name="Species 2",
                           title="Synteny Comparison")
plt.savefig('synteny_dotplot.png', dpi=120, bbox_inches='tight')
plt.close()
print("Saved: synteny_dotplot.png")
```

## 4. Network Visualization

```python
from biosuite.plotting.network_plots import create_ppi_network, plot_network

# Create PPI network
interactions = [
    ('BRCA1', 'TP53', 0.8),
    ('TP53', 'MYC', 0.6),
    ('MYC', 'EGFR', 0.7),
    ('EGFR', 'PTEN', 0.5),
    ('PTEN', 'BRCA1', 0.9),
]

G = create_ppi_network(interactions)
fig = plot_network(G, title="PPI Network")
plt.savefig('ppi_network.png', dpi=120, bbox_inches='tight')
plt.close()
print("Saved: ppi_network.png")
```
